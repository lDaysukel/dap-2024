1. Создать массив 10x10 со случайными значениями, найти минимум и максимум (низкая)

2. Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке

````python

chessboard = np.zeros((8, 8), dtype=int)
chessboard[1::2, ::2] = 1
chessboard[::2, 1::2] = 1

print("Шахматная доска 8x8:")
print(chessboard)
````

3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

````python
z = np.arange(11)

z[(z > 3) & (z < 8)] *= -1

print("Массив после изменения знака:")
print(z)
````

4. Найти среднее значение строки матрицы и вычесть его из каждого элемента строки

````python
z = np.random.rand(5, 10)

row_means = np.mean(z, axis=1, keepdims=True)

z_centered = z - row_means

print("Исходная матрица:")
print(z)
print("\nСредние значения строк:")
print(row_means)
print("\nМатрица после вычитания среднего:")
print(z_centered)
````

6. Дан массив 10x2 представлющий собой точки в декартовой системе координат, преобразовать в полярную

````python
z = np.random.random((10,2))

print("Исходные координаты (x, y):")
print(z)

x, y = z[:, 0], z[:, 1]
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

polar_coords = np.column_stack((r, theta))

print("\nПолярные координаты (r, theta):")
print(polar_coords)
````

7. Заменить максимальный элемент массива на ноль

8. Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз

````python
z = np.ones(10)
i = np.random.randint(0,len(z),20)

print("Исходный массив z:")
print(z)
print("\nМассив индексов i:")
print(i)

unique_indices = np.unique(i)

z[unique_indices] += 1

print("\nМассив z после добавления 1 к элементам с уникальными индексами:")
print(z)
````

9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

````python
z = np.array([1,2,3,4,5])
new_z = np.zeros(len(z) + (len(z) - 1) * 3, dtype=int)
new_z[::4] = z

print("Исходный вектор:")
print(z)
print("\nНовый вектор с тремя нулями между каждым значением:")
print(new_z)
````

10. Найти ближайшее к заданному значению число в заданном массиве

````python
z = np.arange(100) # Число
v = np.random.uniform(0,100) # Массив

nearest_index = np.argmin(np.abs(z - v))

nearest_value = z[nearest_index]

print(f"Случайное число v: {v}")
print(f"Ближайшее число в массиве z: {nearest_value} (индекс: {nearest_index})")
````

**10**.  В чем преимущество массивов NumPy по сравнению с обычными списками.  Чем обусловлено это преимущество?

- Производительность: NumPy использует низкоуровневые оптимизации (C/C++ под капотом), что делает операции с массивами быстрее.

- Удобство: NumPy предоставляет множество встроенных функций для работы с массивами (математические операции, статистика, линейная алгебра и т.д.).

- Экономия памяти: NumPy массивы хранят данные более компактно, чем списки Python.

- Векторизация: Операции применяются ко всему массиву сразу, без необходимости писать циклы.

Это преимущество обусловлено использованием непрерывных блоков памяти и оптимизированных низкоуровневых операций.
